# **Descrição dos dados**

- #### Os dados utilizados neste projeto são registros de escuta de músicas extraídos da minha conta pessoal do Spotify, contendo informações como data e hora de reprodução, nome da faixa, nome do artista, tempo de reprodução em milissegundos e o motivo dela ter sido tocada.

* **ts**: Este campo é um carimbo de data/hora indicando quando a faixa parou de ser tocada em formato UTC (Tempo Universal Coordenado). A ordem é ano, mês e dia, seguida por um carimbo de data/hora no formato de horário militar.

* **ms_played**: Este campo é a duração do streaming, em milésimos de
segundo.

* **master_metadata_track _name**: Este campo é o nome da faixa.

* **master_metadata_album_artist_name**: Este campo é o nome do artista, da banda ou do podcast.

* **reason_start**: Este campo é um valor informando por que a faixa foi
reproduzida (por exemplo, “trackdone”).

    * clickrow: clique na linha
    * trackdone: faixa concluída
    * fwdbtn: botão de avanço
    * backbtn: botão de retrocesso
    * remote: remoto
    * appload: carregamento da faixa

# Importar bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

---
# Filtragem dos dados

```python
df = pd.read_csv('retrospect_2023-2024.csv')
```

```python
# Criando uma nova coluna com valores convertidos de ms para minutos (com 2 casas decimais)
df['min_played'] = (df['ms_played'] / 60000).round(2)

# Exibindo o DataFrame atualizado
df.head()
```

```python
# Verificar as colunas disponíveis
print(df.columns)
```

```python
# Mostrar informações sobre o DataFrame
print(df.info())
```

---
# Determinação do limite de Data
### 1 de Janeiro de 2024 até 20 de Novembro de 2024 (Dia da finalização da coleta dos dados pra Retrospectiva de 2024)

```python
# Converter a coluna 'ts' para datetime
df['ts'] = pd.to_datetime(df['ts']).dt.date

# Definir a data limite: consideramos de 1º de janeiro de 2024 a 20 de novembro de 2024
data_limite_inicio = pd.to_datetime('2024-01-01').date()
data_limite_fim = pd.to_datetime('2024-11-20').date()

# Filtrar as linhas dentro do período
df = df[(df['ts'] >= data_limite_inicio) & (df['ts'] <= data_limite_fim)]

# Verificando o DataFrame após a filtragem
df.head()
```

---
# Quantidade total de minutos escutados no ano

```python
# Calcular o total de minutos
total_minutes = df['min_played'].sum()

print(f"Total de minutos escutados: {total_minutes:.2f} minutos")
```

---
# Dia que mais tocou e a quantidade de minutos ouvidos aquele dia

```python
# Verificando o dia com o maior número de minutos
day_with_most_minutes = df.groupby('ts')['min_played'].sum().idxmax()  # Grupo por 'ts' e somando os minutos
total_minutes_on_day_check = df[df['ts'] == day_with_most_minutes]['min_played'].sum()

print(f"Dia com o maior número de minutos: {day_with_most_minutes}")
print(f"Total de minutos no dia: {total_minutes_on_day_check:.2f} minutos")
```

## Música mais ouvida naquele dia e quantidade de vezes

```python
# Filtrando as linhas do DataFrame para o dia com o maior número de minutos
df_day_most_minutes = df[df['ts'] == day_with_most_minutes]

# Encontrando a música mais ouvida no dia
most_played_song = df_day_most_minutes.loc[df_day_most_minutes['min_played'].idxmax()]

# Contando quantas vezes a música foi tocada no dia com o maior número de minutos
play_count = df_day_most_minutes[df_day_most_minutes['master_metadata_track_name'] == most_played_song['master_metadata_track_name']].shape[0]

print(f"Música mais ouvida no dia {day_with_most_minutes}: {most_played_song['master_metadata_track_name']}")
print(f"Artista: {most_played_song['master_metadata_album_artist_name']}")
print(f"Minutos tocados: {most_played_song['min_played']:.2f}")
print(f"A música '{most_played_song['master_metadata_track_name']}' foi tocada {play_count} vezes")
```

```python
df['ts'] = pd.to_datetime(df['ts'])

# Definindo o "days around" (15 dias antes e 15 dias depois)
days_around = 15

# Obtenção do intervalo de datas
date_range = pd.date_range(start=day_with_most_minutes - pd.Timedelta(days=days_around), 
                           end=day_with_most_minutes + pd.Timedelta(days=days_around), 
                           freq='D')

# Contagem de minutos por dia no intervalo
daily_minutes = df[df['ts'].dt.date.isin(date_range.date)].groupby('ts')['min_played'].sum()

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(daily_minutes.index, daily_minutes.values, marker='o', color='b', linestyle='-')

# Configurações do gráfico
plt.title(f"Minutos Escutados nos 15 Dias Antes e 15 Dias Depois do Dia com Mais Minutos")
plt.xlabel("Data")
plt.ylabel("Minutos Escutados")
plt.xticks(rotation=45)
plt.grid(True)

# Formatando os ticks da data
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.axvline(x=day_with_most_minutes, color='r', linestyle='--', label=f"Dia Mais Escutado ({day_with_most_minutes})")
plt.legend()
plt.tight_layout()
plt.show()
```

## Músicas tocadas em 13 de Abril de 2024

```python
# Filtrando as músicas tocadas no dia 13 de abril
day_of_interest = '2024-04-13'
df_april_13 = df[df['ts'] == day_of_interest]

# Agrupar por nome da música e somar o tempo total de reprodução
df_top_musicas_april_13 = df_april_13.groupby('master_metadata_track_name')['min_played'].sum().reset_index()

# Ordenar em ordem decrescente pelo tempo total de reprodução
df_top_musicas_april_13 = df_top_musicas_april_13.sort_values(by='min_played', ascending=False)

# Selecionar apenas as 15 músicas mais tocadas
df_top_musicas_april_13 = df_top_musicas_april_13.head(15)

# Plotando as músicas
plt.figure(figsize=(10, 6))
plt.bar(df_top_musicas_april_13['master_metadata_track_name'], df_top_musicas_april_13['min_played'], color='skyblue')

# Adicionando título e rótulos aos eixos
plt.title(f'Top 15 Músicas Tocadas em 13 de Abril de 2024 (Minutos Escutados)')
plt.xlabel('Nome da Música')
plt.ylabel('Minutos Escutados')

# Rotacionando os nomes das músicas no eixo x
plt.xticks(rotation=45, ha='right')

# Mostrando o gráfico
plt.tight_layout()
plt.show()
```

---
# Top 5 músicas mais tocadas

```python
# Filtrar apenas as músicas que foram reproduzidas (ms_played > 0)
df_reproduzidas = df[df['min_played'] > 0]

# Agrupar por nome da música e somar o tempo total reproduzido
df_top_musicas = df_reproduzidas.groupby('master_metadata_track_name').agg({'min_played': 'sum'}).reset_index()

# Ordenar em ordem decrescente pelo tempo total reproduzido
df_top_musicas = df_top_musicas.sort_values(by='min_played', ascending=False)

# Selecionar as 10 músicas mais tocadas
df_top_5 = df_top_musicas.head(5)

# Exibir o Top 10
df_top_5
```

```python
# Configuração do tamanho do gráfico
plt.figure(figsize=(10, 6))

# Gráfico de barras
plt.bar(df_top_5['master_metadata_track_name'], df_top_5['min_played'], color='skyblue')  # Convertendo para minutos

# Adicionando título e rótulos aos eixos
plt.title('Top 5 Músicas Mais Tocadas (Minutos Escutados)')
plt.xlabel('Nome da Música')
plt.ylabel('Minutos Escutados')

# Rotacionando os nomes das músicas no eixo x
plt.xticks(rotation=45, ha='right')

# Mostrando o gráfico
plt.tight_layout()
plt.show()
```

---
# Música mais tocada

```python
# Filtrando apenas as linhas onde a música mais ouvida foi reproduzida
most_played_song = df_top_5.iloc[0]['master_metadata_track_name']  # Pegando o nome da música mais tocada

# Contando os valores de reason_start para a música mais tocada
count_reason_start = df[df['master_metadata_track_name'] == most_played_song]['reason_start'].value_counts()

# Valores da música mais tocada
reason_counts = df[df['master_metadata_track_name'] == most_played_song]['reason_start'].value_counts()

# Somar todos os valores
total_reproducoes = reason_counts.sum()

print(f'Música mais tocada: {most_played_song}')
print("Total de reproduções da música mais tocada:", total_reproducoes)
print(count_reason_start)
```

```python
# Pegando a música mais tocada e suas informações
most_played_song = df_top_5.iloc[0]['master_metadata_track_name']

# Obtendo as informações de tempo de reprodução para essa música
total_minutes_played = df[df['master_metadata_track_name'] == most_played_song]['min_played'].sum()

print(f"Música mais tocada: {most_played_song}")
print(f"Total de minutos tocados ao longo do ano: {total_minutes_played:.2f} minutos")
```

```python
# Agrupando os dados por mês
df['month'] = pd.to_datetime(df['ts']).dt.to_period('M')
monthly_played = df.groupby(['month', 'master_metadata_track_name'])['min_played'].sum().reset_index()

# Filtando apenas para a música mais tocada
most_played_song = df_top_5.iloc[0]['master_metadata_track_name']
monthly_played = monthly_played[monthly_played['master_metadata_track_name'] == most_played_song]

# Configuração do gráfico
plt.figure(figsize=(12, 6))

# Gráfico de linhas
plt.plot(monthly_played['month'].astype(str), monthly_played['min_played'], marker='o', color='b')

# Adicionando título e rótulos aos eixos
plt.title(f'Evolução de Minutos Reproduzidos ao Longo do Ano para a Música "{most_played_song}"')
plt.xlabel('Mês')
plt.ylabel('Minutos Reproduzidos')
plt.xticks(rotation=45)

# Mostrando o gráfico
plt.tight_layout()
plt.show()
```

# Artista mais escutado

```python
# Agrupando por artista e somando os minutos de reprodução
artist_with_most_minutes = df.groupby('master_metadata_album_artist_name')['min_played'].sum().idxmax()  # Artista com o maior total de minutos
total_minutes_by_artist = df.groupby('master_metadata_album_artist_name')['min_played'].sum().max()

print(f"Artista que você mais ouviu: {artist_with_most_minutes}")
print(f"Total de minutos escutados desse artista: {total_minutes_by_artist:.2f} minutos")
```

## Artista mais escutado (Gráfico por semana)

```python
# Agrupando por semana e somando os minutos de reprodução
artist_name = artist_with_most_minutes
artist_data = df[df['master_metadata_album_artist_name'] == artist_name]

# Convertendo a coluna 'ts' para datetime
artist_data['ts'] = pd.to_datetime(artist_data['ts'])

# Agrupando por semanas e somando os minutos
weekly_play_minutes = artist_data.resample('W', on='ts')['min_played'].sum().reset_index()

# Configuração do gráfico
plt.figure(figsize=(10, 6))

# Plotando os dados
plt.plot(weekly_play_minutes['ts'], weekly_play_minutes['min_played'], marker='o', color='blue')

# Adicionando título e rótulos aos eixos
plt.title(f'Minutos Escutados por Semana - {artist_name}')
plt.xlabel('Semana')
plt.ylabel('Minutos Escutados')

# Rotacionando os rótulos do eixo x
plt.xticks(rotation=45)

# Mostrando o gráfico
plt.tight_layout()
plt.show()
```

## Artista mais escutado (Gráfico por mês)

```python
# Agrupando os dados por mês
artist_name = artist_with_most_minutes
artist_data = df[df['master_metadata_album_artist_name'] == artist_name]

# Convertendo a coluna 'ts' para datetime
artist_data['ts'] = pd.to_datetime(artist_data['ts'])

# Agrupando por meses e somando os minutos
monthly_play_minutes = artist_data.resample('M', on='ts')['min_played'].sum().reset_index()

# Configuração do gráfico
plt.figure(figsize=(12, 6))

# Plotando os dados
plt.plot(monthly_play_minutes['ts'], monthly_play_minutes['min_played'], marker='o', color='blue')

# Adicionando título e rótulos aos eixos
plt.title(f'Minutos Escutados por Mês - {artist_name}')
plt.xlabel('Mês')
plt.ylabel('Minutos Escutados')

# Rotacionando os rótulos do eixo x
plt.xticks(rotation=45)

# Mostrando o gráfico
plt.tight_layout()
plt.show()
```
